```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow  # Importing the function specific to Colab for displaying images

# Load an image from file
image_path = '/content/sarinakasaiyan.png'  # Specify the path to your image file
loaded_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load the image in grayscale

# Check if the image was loaded successfully
if loaded_image is None:
    print("Error: Unable to load image. Please check the path.")
else:
    # Display the loaded image
    cv2_imshow(loaded_image)

# Create a blank black image of size 300x300
image = np.zeros((300, 300), dtype=np.uint8)

# Draw a white circle at a new position
cv2.circle(image, (200, 100), 50, (255), -1)  # Circle with center (200,100) and radius 50

# Draw a white rectangle at a new position
cv2.rectangle(image, (50, 200), (150, 300), (255), -1)  # Rectangle with corners (50,200) and (150,300)

# Add noise to the image
noise = np.random.randint(0, 2, image.shape) * 255  # Create binary noise
noisy_image = cv2.bitwise_or(image, noise.astype(np.uint8))  # Combine original image with noise

# Save the original and noisy images
cv2.imwrite('binary_image_updated.png', image)         # Save the original binary image
cv2.imwrite('noisy_image_updated.png', noisy_image)   # Save the noisy image

# Display the images
cv2_imshow(image)          # Show the original binary image
cv2_imshow(noisy_image)    # Show the noisy image

# Create a new blank image for comparison
new_image = np.zeros((300, 300), dtype=np.uint8)

# Draw a white triangle at a new position
points = np.array([[150, 50], [100, 150], [200, 150]], np.int32)  # Coordinates of the triangle
cv2.fillPoly(new_image, [points], (255))  # Fill the triangle with white color

# Draw a white rectangle at a new position
cv2.rectangle(new_image, (20, 200), (100, 250), (255), -1)  # Rectangle with corners (20,200) and (100,250)

# Save and display the new image
cv2.imwrite('new_shape_image.png', new_image)        # Save the new shape image
cv2_imshow(new_image)          # Show the new shape image
```
